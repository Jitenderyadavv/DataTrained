```python
#lets use encoding the object datatype
numerics = ['int8','int16','int32','int64','float16','float32','float64']

categorical_column =[]
features = data.columns.values.tolist()

for col in features:
    if data[col].dtype in numerics:
        continue
    categorical_column.append(col)
    
categorical_column
```
